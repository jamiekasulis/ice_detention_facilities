```python
# Facilities Main information
main_df = pd.read_excel("https://github.com/jamiekasulis/ice_detention_facilities/blob/master/data/raw_excel_facilities_main.xlsx?raw=true",
                                   skiprows=8)
```

### Begin regional analysis: CT

```python
# grab CT and new england facilities
new_england = ['CT', 'MA', 'ME', 'VT', 'RI', 'NH']
ct_facilities = main_df[main_df['State'] == 'CT']
ne_facilities = main_df[main_df['State'].isin(new_england)]
```

Of CT facilities...
* None are monitored for compliance with ICE detention standards.
* __Only Harhold was in use in FY17.__ Only 3 people were detained here in FY17. All other detention centers in CT were unused in FY17, according to this data.
